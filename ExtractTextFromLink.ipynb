```python
import urllib.request
import urllib.parse
import re
```

```python
url = "https://deepak-mandal.github.io"

values = {'s': 'basics',
         'submit': 'search'}


```

```python
data = urllib.parse.urlencode(values)
data = data.encode('ascii')#('utf-8')
```

```python
req = urllib.request.Request(url, data)
```

```python
#import urllib.request
resp = urllib.request.urlopen(req)
#resp = urllib.request.urlopen('https://deepak-mandal.github.io/')
```

```python
respData = resp.read()
```

```python
print(respData)
```

```python
paragraphs = re.findall(r'<p>(.*?)</p>', str(respData))
```

```python
for eachP in paragraphs:
    print(eachP)
```
