```python
#Creating bunch of sentences
raw_docs = ["I am writing some very basic english sentence",
           "I'm just writting it for the demo pURPOSE to make audience understand the basic .",
           "The point is to _learn HOW it works_ on #simple # data."]
```

```python
filename = "text.txt"
raw_docs = open(filename).read()
print(raw_docs)
```

```python
#Importing nltk package
import nltk
#nltk.download
```

Converting to lower case

```python
import string
raw_docs = [doc.lower() for doc in raw_docs]
print(raw_docs)
```

Tokenization 

```python
#word tokenize :- breaking the sentence into words, or if you have multiple sentences then taking you sentence separate
from nltk.tokenize import word_tokenize
tokenized_docs = [word_tokenize(doc) for doc in raw_docs]
print(tokenized_docs)

print("##########################################################")


#sentence tokenization

from nltk.tokenize import sent_tokenize
sent_token = [sent_tokenize(doc) for doc in raw_docs]
print(sent_token)
```

Punctuation removal

```python
import re    #Regular expression
#regular expression funct
regex = re.compile('[%s]' % re.escape(string.punctuation))    

tokenized_docs_no_punctuation = []

for review in tokenized_docs:
    new_review = []
    for token in review:
        new_token = regex.sub(u'', token)
        if not new_token == u'':
            new_review.append(new_token)
            
    tokenized_docs_no_punctuation.append(new_review)
    
print(tokenized_docs_no_punctuation)

            
```

Removing stopwords

```python
# cleaning text of stopwords
from nltk.corpus import stopwords

tokenized_docs_no_stopwords = []

for doc in tokenized_docs_no_punctuation:
    new_term_vector = []
    for word in doc:
        if not word in stopwords.words('english'):
            new_term_vector.append(word)
    
    tokenized_docs_no_stopwords.append(new_term_vector)

print(tokenized_docs_no_stopwords)
```

Stemming - just removing from the end such as ing , es etc to get the english word 
and Lemmantization - converted to roo words

```python
#stemming and Lemmatization
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer

porter = PorterStemmer()
wordnet = WordNetLemmatizer()

preprocessed_docs = []

for doc in tokenized_docs_no_stopwords:
    final_doc = []
    for word in doc:
        #final_doc.append(porter.stem(word))
        final_doc.append(wordnet.lemmatize(word))
        
    preprocessed_docs.append(final_doc)
    
print(preprocessed_docs)

```
