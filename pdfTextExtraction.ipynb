```python
import PyPDF2 as pdf
```

```python
file = open('nlp.pdf', 'rb')    #rb: read binary
```

```python
file
```

```python
pdf_reader = pdf.PdfFileReader(file)
```

```python
print(pdf_reader.documentInfo)
```

```python
help(pdf_reader)
```

```python
pdf_reader.getIsEncrypted()
```

```python
pdf_reader.getNumPages()
```

```python
page1=pdf_reader.getPage(10)
```

```python
#page1    #in binary form
page1.extractText()
```

```python
print(page1.extractText())
```

```python
page2 = pdf_reader.getPage(1)
```

```python
page2.extractText()
```

```python
print(page2.extractText())
```

```python
##Append Write or Merge PDFs
```

```python
pdf_writer = pdf.PdfFileWriter()
```

```python
pdf_writer.addPage(page1)
```

```python
output = open('Pages.pdf',  'wb')
pdf_writer.write(output)
output.close()
```

```python
#multiple pages
pdf_writer = pdf.PdfFileWriter()

pdf_writer.addPage(page1)
pdf_writer.addPage(page2)


output = open('Pages.pdf',  'wb')
pdf_writer.write(output)
output.close()
```

```python
import PyPDF2
a = PyPDF2.PdfFileReader('nlp.pdf')

totalPages = a.getNumPages()

str=""
for i in range(totalPages):
    str += a.getPage(i).extractText()
    

with open("text.txt", "w", encoding = "utf-8") as f:
    f.write(str)
```
